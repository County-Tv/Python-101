# Break Clause

```
House = input("what type of house do you want: ")
Apartment = 1
for Rooms in House:
  print("The Letter" ,Apartment ," is", Rooms)
  Apartment +=1
  if Apartment > 6:
    print("This name is to long")
    break
```
