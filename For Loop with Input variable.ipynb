# For Loop With Input

```
House = input("What type of House do you have: ")
Characters = 1
for Letters in House:
  print("The first number ",  Characters , " in the House input is" , Letters)
  Characters += 1
  if Characters > 6:
    print(" this list is to long")
    break
```
